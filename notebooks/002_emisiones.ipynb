```python
import pandas as pd
import plotly.express as px
```

```python
f = "../data/owid-energy-data.csv"
energia = pd.read_csv(f)
energia.head()
```

```python
columnas = energia.columns
for columna in columnas:
    print(columna)
```

```python
energia.iso_code.unique()
```

```python

lista_paises = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Costa Rica', 
                'Cuba', 'Dominican Republic', 'Ecuador', 'Guatemala', 'Honduras', 
                'Mexico',  'Panama', 'Paraguay', 'Peru', 
                'El Salvador', 'Uruguay', 'Venezuela', 'World'] 

paises = energia[energia["country"].isin(lista_paises)]


fig = px.line(paises, x="year", y="greenhouse_gas_emissions", hover_name="country", color="country",
             range_x=[2000,2024],
             labels={
                     "year": "",
                 })  

# fig.update_layout(
#     yaxis_title="EnergÃ­a Primaria [TWh]"
# )
fig.show()
```
